```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,10)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
from matplotlib.lines import lineStyles
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,color='m', linestyle=":", linewidth=6)
plt.show()
```

```python
x= np.arange(1,11)
y1= 2*x
y2=3*x
```

```python
plt.plot(x,y1,color='orange', linestyle='--', linewidth=2)
plt.plot(x,y2,color='red', linestyle=":",linewidth=2)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["y=2x","y=3x"])
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color='orange', linestyle='--', linewidth=2)
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,y2,color='red', linestyle=":",linewidth=2)

plt.grid(True)
plt.show()
```

```python
student = {"jungkook":97, "jimin":87, "thv":56}

names = list(student.keys())
values= list(student.values())

plt.bar(names,values)
plt.show()
```

```python
x=[10,20,20,80,90,78,56,34]
a=[3,5,7,8,9,5,13,45]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.scatter(x,a,color="r", marker="*",s=100)
plt.show()
```

```python
x=[10,20,20,80,90,78,56,34]
a=[3,5,7,8,9,5,13,45]
c=[1,2,3,5,7,9,0,5]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.scatter(x,a,color="r", marker="*",s=100)
plt.scatter(x,c,color="m", marker="d",s=200)

plt.show()
```

```python
x=[10,20,20,80,90,78,56,34]
a=[3,5,7,8,9,5,13,45]
c=[1,2,3,5,7,9,0,5]

plt.subplot(1,2,1)
plt.scatter(x,a,color="r", marker="*",s=10)

plt.subplot(1,2,2)
plt.scatter(x,c,color="m", marker="d",s=20)

plt.show()
```
